```python
arr_txt=arr_txt.dropna()
arr_txt.count()
```

```python
#plots
import seaborn as sns
```

```python
#scatter plot
plt.figure(figsize=(8,5))
sns.scatterplot(x="Pclass",y="PassengerId",data=arr_txt,palette="deep")
plt.title("SCATTER PLOT",fontsize=14)
plt.show()
```

```python
#bar plot
plt.figure(figsize=(8,5))
sns.barplot(x="Ticket",y="Fare",data=arr_txt,palette="muted")
plt.title("BAR PLOT",fontsize=14)
plt.show()
```

```python
#box plot
plt.figure(figsize=(10,7))
sns.boxplot(x="SibSp",y="Fare",data=arr_txt,palette="muted")
plt.title("BOX PLOT",fontsize=12)
plt.show()
```

```python
#density plot
plt.figure(figsize=(8,5))
sns.kdeplot(x="PassengerId",y="SibSp",data=arr_txt,palette="muted")
plt.title("DENSITY PLOT",fontsize=14)
plt.show()
```

```python
#violin plot
plt.figure(figsize=(8,5))
sns.violinplot(x="Ticket",y="Fare",data=arr_txt,palette="pastel")
plt.title("VIOLIN PLOT",fontsize=14)
plt.show()
```

```python
#line plot
plt.figure(figsize=(8,5))
sns.lineplot(x="Ticket",y="Embarked",data=arr_txt,palette="colorblind")
plt.title("LINE PLOT",fontsize=14)
plt.show()
```

```python
#histogram
plt.figure(figsize=(8,5))
sns.histplot(x="Ticket",y="Fare",data=arr_txt,palette="muted")
plt.title("HISTOGRAM",fontsize=14)
plt.show()
```

```python
#count plot
plt.figure(figsize=(8,5))
sns.countplot(x="Parch",data=arr_txt,palette="muted")
plt.title("COUNT PLOT",fontsize=14)
plt.show()
```

```python
#pie chart
plt.figure(figsize=(8,5))
plt.pie(x="Age",data=arr_txt)
plt.title("PIE PLOT",fontsize=14)
plt.show()
```
